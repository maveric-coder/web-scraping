```python
import requests
r = requests.get(url, headers=headers)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text,'html.parser')
```

```python
soup
```

```python
results=soup.find_all('div', attrs={'class':'s-include-content-margin s-border-bottom'})
```

```python
first_result = results[0]
```

```python
##          title
first_result.find('span', attrs={'class':'a-size-medium a-color-base a-text-normal'}).text
```

```python
##                 url

'https://www.amazon.in'+first_result.find('a')['href']
```

```python
##    price

first_result.find_all('span', attrs={'class':'a-offscreen'})[0].text.strip()
```

```python
#### delivery charges

first_result.find_all('span', attrs={'aria-label':'FREE Delivery by Amazon'})[0].text.strip()
```

```python
####  ratings

first_result.find_all('span', attrs={'class':'a-icon-alt'})[0].text.strip()
```

```python
records = []
for result in results:
    title = result.find('span', attrs={'class':'a-size-medium a-color-base a-text-normal'}).text
    price = result.find_all('span', attrs={'class':'a-offscreen'})[0].text.strip()
    delivery_charges = result.find_all('span', attrs={'aria-label':'FREE Delivery by Amazon'})[0].text.strip()
    rating=result.find_all('span', attrs={'class':'a-icon-alt'})[0].text.strip()
    url='https://www.amazon.in'+result.find('a')['href']
    records.append((title,price,delivery_charges, rating, url))
```

```python
len(records)
```

```python
records[0:3]
```

```python
import pandas as pd
df = pd.DataFrame(records, columns=['title','price','delivery_charges', 'rating', 'url'])
```

```python
df.to_csv('amazon_smartphones.csv', index=False, encoding='utf-8')
```
